### import Necessary Libraries

```python
import pandas as pd
import os
```

#### merging 12 months of sales data into a single Csv file

```python
df = pd.read_csv("C:/Users/megha/OneDrive/pandas/Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/Sales_April_2019.csv")

files=[file for file in os.listdir("C:/Users/megha/OneDrive/pandas/Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data")]

all_months_data=pd.DataFrame()

for file in files:
    df=pd.read_csv("C:/Users/megha/OneDrive/pandas/Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/"+file)
    all_months_data=pd.concat([all_months_data,df])
```

#### Read DataFrame

```python
all_months_data.head()
```

## clean up the data!

#### Drop rows of NAN

```python
nan_data= all_months_data[all_months_data.isna().any(axis=1)]
nan_data.head()

all_months_data=all_months_data.dropna(how='all')
all_months_data.head()
```

#### Find 'Or' and Delete it

```python
all_months_data= all_months_data[all_months_data['Order Date'].str[0:2] !='Or']
```

#### convert columns to the correct type

```python
all_months_data['Quantity Ordered'] = pd.to_numeric(all_months_data['Quantity Ordered']) #make int
all_months_data['Price Each'] = pd.to_numeric(all_months_data['Price Each']) #make float
all_months_data.head()
```

## Augment data with additional columns

#### Add month column to the data

```python
all_months_data['Month'] = all_months_data['Order Date'].str[0:2]
all_months_data['Month'] = all_months_data['Month'].astype('int32')
all_months_data.head()
```

#### Add a sales column

```python
all_months_data['Sales'] = all_months_data["Quantity Ordered"] * all_months_data["Price Each"]
all_months_data.head()
```

#### Add a city column

```python
def get_city(address):
    return address.split(',')[1].strip()  # Use .strip() to remove leading/trailing whitespace

def get_state(address):
    state_part = address.split(',')[2].strip()  # Use .strip() to remove leading/trailing whitespace
    return state_part.split(' ')[0]  # Get the first element of the split list, which should be the state name

all_months_data["City"] = all_months_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")

all_months_data.head()
```

# Questions

#### question 1: What was the best month for sales?How much was earned that month?

```python
col=['Sales']
result=all_months_data.groupby('Month')[col].sum()
result
```

```python
import matplotlib.pyplot as plt

months=range(1,13)

plt.bar(months,result['Sales'])
plt.xticks(months)
plt.ylabel('sales')
plt.xlabel('Month')
plt.show()
```

#### Question 2: Which city has the highest number of sales

```python
results=all_months_data.groupby('City')[['Sales']].sum()
results
```

```python
import matplotlib.pyplot as plt

cities=all_months_data['City'].unique()

plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation = 'vertical',size=8)
plt.ylabel('sales')
plt.xlabel('Month')
plt.show()
```

#### Question 3:what products are most often sold together?

```python
df=all_months_data[all_months_data['Order ID'].duplicated(keep=False)]
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df=df[['Order ID','Grouped']].drop_duplicates()

df.head(100)
```

```python
from itertools import combinations 
from collections import Counter

count=Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
    print(key,value)
```

#### Question 4: What product sold the most?

```python
product_group = all_months_data.groupby('Product')

quantity_ordered = product_group.sum()['Quantity Ordered']

products=[product for product,df in product_group]

plt.bar(products,quantity_ordered)

plt.xticks(products,rotation='vertical',size=8)
plt.show()
```
